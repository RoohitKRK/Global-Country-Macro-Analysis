# Data Preprocessing

```python
data=data.drop(['Country','Abbreviation','Currency-Code','Gasoline Price','Minimum wage','Official language','Population: Labor force participation (%)','Tax revenue (%)','Latitude','Longitude','Largest city','Capital/Major City','Calling Code'],axis=1)
print(data)
```

```python
data.info()
```

```python
df_cleaned = data.replace({'\$': '', ',': '', '%': ''}, regex=True)
df_cleaned =df_cleaned.apply(pd.to_numeric, errors='coerce')
data=df_cleaned
print(data)
```

```python
categorical_columns = data.select_dtypes(include=['object']).columns
continuous_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Replace null values in categorical columns with the mode
for column in categorical_columns:
    mode_value = data[column].mode()[0]
    data[column].fillna(mode_value, inplace=True)

# Replace null values in continuous columns with the mean
for column in continuous_columns:
    mean_value = data[column].mean()
    data[column].fillna(mean_value, inplace=True)
```

# Multicollinearity

```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd # Make sure pandas is imported

def calc_vif(x):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = x.columns
    vif["VIF"] = [variance_inflation_factor(x.values, i).round(1) for i in range(x.shape[1])]

    return(vif)

# Now call calc_vif with x_train:
calc_vif(x)
```
